```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
import pandas as pd
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('data_out.csv')
X=df['text']
y=df['category']

class_counts = df['category'].value_counts()
print(class_counts)
```

Feature Extraction

```python
vectorizerTF = TfidfVectorizer()
X_vec_tf = vectorizerTF.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_vec_tf, y, test_size=0.2, random_state=42, shuffle=True)
```

Logistic regression

```python
modelL = LogisticRegression(solver='lbfgs', max_iter=1000, class_weight='balanced')
modelL.fit(X_train, y_train)
y_pred = modelL.predict(X_test)
```

```python
print("Accuracy: " , accuracy_score(y_test, y_pred))
```

Naive Bayes

```python
modelNB = MultinomialNB()
modelNB.fit(X_train, y_train)
y_pred = modelNB.predict(X_test)
```

```python
print("Accuracy: " , accuracy_score(y_test, y_pred))
```

```python
new_texts = [
    "nali nihari",
    "petrol",
    "Books"
]
X_new = vectorizerTF.transform(new_texts)
y_pred_new = modelL.predict(X_new)

print("Predicted categories:", y_pred_new)
```
